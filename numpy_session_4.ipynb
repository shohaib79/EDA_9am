**covariance**

```python
x=range(1,11)
y=range(100,110)
len(x),len(y)
```

```python
list(x),list(y)
```

```python
arr1=np.array(x)
arr1_mean=np.mean(arr1)
arr11=arr1-arr1_mean

arr2=np.array(y)
arr2_mean=np.mean(arr2)
arr22=arr2-arr2_mean

arr3=arr11*arr22

arr4=np.sum(arr3)

arr4/9
```

```python
arr1=np.array(x)
arr2=np.array(y)

np.sum((arr1-np.mean(arr1))*(arr2-np.mean(arr2)))/9
```

**corelation coeeficaiant**

```python
arr1=np.array(x)
arr1_mean=np.mean(arr1)
arr11=arr1-arr1_mean

arr2=np.array(y)
arr2_mean=np.mean(arr2)
arr22=arr2-arr2_mean

Nr=np.sum(arr11*arr22)

a=np.sum(np.square(arr11))
b=np.sum(np.square(arr22))
Dr=np.sqrt(a*b)

Nr/Dr
```

**cosine similarity**

```python
x=[1,2,3]
y=[4,5,6]
arr1=np.array(x)
arr2=np.array(y)
Nr=np.dot(arr1,arr2)
x1=norm(arr1)
x2=norm(arr2)
Dr=x1*x2
Nr/Dr
```
