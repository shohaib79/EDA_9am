```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python

import warnings
warnings.filterwarnings('igonre')
model=SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences=[
    "I like apples.",
    "I enjoy eating apples.",
    "I drive a motorcycle."
]
```

```python
embeddings = model.encode(sentences)
```

```python
v1=embeddings[0].reshape(-1,1) #I like apples
v2=embeddings[1].reshape(-1,1) #I like apples
v3=embeddings[2] .reshape(-1,1)#
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(v1,v2)
```

```python
def cosine_sim(a,b):
    return np.dot(a,b)/(np.linalg.norm(a) * np.linalg.norm(b))
```

```python
v1=embeddings[0]
v2=embeddings[1]
Nr=np.dot(v1,v2)
norm_v1=np.linalg.norm(v1)
norm_v2=np.linalg.norm(v2)
Dr=norm_v1*norm_v2
cosine_sim(v1,v2),Nr/Dr
```
